```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([3, 4, 5, 6, 7])      
y = np.array([350, 450, 550, 650, 750]) 
```

```python
m = len(x)
```

```python
w = 0    
b = 0 
```

```python
alpha = 0.01
epochs = 1000
```

```python
for epoch in range(epochs):

    y_pred = w * x + b
    error = y_pred - y
    dw = (1/m) * np.dot(error, x)
    db = (1/m) * np.sum(error)
    w = w - alpha * dw
    b = b - alpha * db
    if epoch % 100 == 0:
        loss = (1/(2*m)) * np.sum(error ** 2)
        print(f"Epoch {epoch}: Loss = {loss:.2f}, w = {w:.2f}, b = {b:.2f}")
```

```python
print("\nFinal values:")
print(f"Weight (w): {w}")
print(f"Bias (b): {b}")
y_hat = w * x + b
print(f"Predictions: {y_hat}")
```
